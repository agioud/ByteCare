```python
df = pd.read_csv('alzheimers_disease_data.csv', sep=',').drop(columns='DoctorInCharge')
```

Demographic Details:
- Age : from 60 to 90 years
- Gender : 0=Male and 1=Female.
- Ethnicity : 0=Caucasian, 1=African American, 2=Asian 3=Other
- EducationLevel : 0=None, 1=High School, 2=Bachelor's, 3=Higher

Lifestyle Factors:
- BMI(Body Mass Index) :ranging from 15 to 40.
- Smoking(Smoking status):0 indicates 0=No, 1=Yes.
- AlcoholConsumption(Weekly): from 0 to 20.
- PhysicalActivity(hours/week): from 0 to 10.
- DietQuality: from 0 to 10.
- SleepQuality: from 4 to 10.

Medical history :
- FamilyHistoryAlzheimers: 0=No, 1=Yes.
- CardiovascularDisease: 0=No, 1=Yes.
- Diabetes: 0=No, 1=Yes.
- Depression: 0=No, 1=Yes.
- HeadInjury: 0=No, 1=Yes.
- Hypertension: 0=No, 1=Yes.

Clinical Measurements
- SystolicBP(Systolic blood pressure): from 90 to 180 mmHg.
- DiastolicBP(Diastolic blood pressure): from 60 to 120 mmHg.
- CholesterolTotal(cholesterol levels): from 150 to 300 mg/dL.
- CholesterolLDL(Low-density lipoprotein cholesterol levels): from 50 to 200 mg/dL.
- CholesterolHDL(High-density lipoprotein cholesterol levels): from 20 to 100 mg/dL.
- CholesterolTriglycerides(Triglycerides levels): from 50 to 400 mg/dL.

Cognitive and Functional Assessments
- MMSE(Mini-Mental State Examination score): from 0 to 30. Lower scores indicate cognitive impairment.
- FunctionalAssessment(Functional assessment score): from 0 to 10. Lower scores indicate greater impairment.
- MemoryComplaints(Presence of memory complaints): 0=No, 1=Yes.
- BehavioralProblems(Presence of behavioral problems): where 0=, 1=Yes.
- ADL(Activities of Daily Living score): from 0 to 10. Lower scores indicate greater impairment.

Symptoms
- Confusion(Presence of confusion): 0=No, 1=Yes.
- Disorientation(Presence of disorientation): 0=No, 1=Yes.
- PersonalityChanges(Presence of personality changes): 0=No, 1=Yes.
- DifficultyCompletingTasks(Presence of difficulty completing tasks): 0=No, 1=Yes.
- Forgetfulness(Presence of forgetfulness): 0=No, 1=Yes.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
import pickle

X = df[['Age','Gender','EducationLevel','Smoking','AlcoholConsumption','PhysicalActivity','DietQuality','SleepQuality','FamilyHistoryAlzheimers',
        'CardiovascularDisease','HeadInjury', 'CholesterolTotal','MMSE','MemoryComplaints',
        'Confusion','Disorientation','PersonalityChanges','DifficultyCompletingTasks','Forgetfulness']]
y = df['Diagnosis']

# Diviser les donnes
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Normaliser les donnes    
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Entraner le modle
model = LogisticRegression()
model.fit(X_train_scaled, y_train)

# Sauvegarder le modle et le scaler
with open('model.pkl', 'wb') as f:
    pickle.dump(model, f)

with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)
```
